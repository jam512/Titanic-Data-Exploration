```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1= pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/train.csv")
data1.head()
```

```python
data2= pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/test.csv")
data2.head()
```

# Concatenate the datasets.

```python
frame=[data1,data2]
result=pd.concat(frame, sort=False)
result
```

# Remove NaN values from the column ‘Survived’.

```python
result.dropna(subset=['Survived'])
```

```python
df=pd.DataFrame(result)
df
```

# Create a new column 'Has_Cabin' that tells you whether passengers had a cabin or not.

```python
df['Has_cabin']=df['Cabin'].notnull()
df
```

# Create a new column “Title”, which will contain the titles of the passengers such as Mr., Mrs., etc. There should be five titles: Mr, Mrs, Master, Miss, and Special.

```python
conditions = [
    (df['Sex'] == 'male') & (df['Age'] <= 18.0),
    (df['Sex'] == 'male') & (df['Age'] >= 18.0),
    (df['Sex'] == 'female') & (df['Age']<=18.0)]
choices = ['Master','Mr','Miss']
df['Titles'] = np.select(conditions, choices, default='Mrs')
df.tail()
```

```python
df1=result.dropna(subset=['Survived'])
df1
```

```python
df.mean()
```

# Impute the missing values using .fillna() method.

```python
df1.fillna(df.mean())
```

```python
df1
```

```python
df2=df1.reindex(columns=[*df1.columns.tolist(), 'Categorical_age', 'Categorical_fare'], fill_value=0)
df2
```

```python
df2
```

```python

df3=df2.drop([ 'Cabin','Name','Ticket','Age','PassengerId','Fare','Parch','SibSp'], axis=1)
df3
```

# Transform all variables into numeric ones using get_dummies(). 

```python
dum=pd.get_dummies(df3)
dum.head()
```

```python
x = dum.iloc[:,1:].values
x
```

```python
x.shape
```

```python
y=dum.iloc[:,0].values
y
```

```python
y.shape
```

```python
y=y.reshape(891,1)
y
```

# Now we need training set and testing set. Use train_test_split() to split your dataset.                                                                                                                        X(Features) will be the dataset values without the column “Survived” and y(Targets) will be your “Survived” column values.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
print (x_train.shape, y_train.shape)
```

```python
y=y_train.reshape(623,1)
y
```

```python

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python

classifier = LogisticRegression(random_state = 0)
classifier.fit(x_train, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(x_test)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```
